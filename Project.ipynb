```python
# Write your answer/code here
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
import contractions

nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()
```

```python
csv_file = 'movies.csv'
movies_df = pd.read_csv(csv_file)

def normalize_text(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)  # removing numbers
    # text = re.sub(r'[^\w\s]', '', text) # removing punctuation
    text = contractions.fix(text) 
    text = re.sub(r"[^\w\s']", "", text)
    text = re.sub(r"'", " ",text)
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)  # Remove special characters
    text = re.sub(r'\s+', ' ', text).strip() # removing extra whitespace

    words = word_tokenize(text)

    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]

    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]

    normalized_text = ' '.join(lemmatized_words)
    
    return normalized_text


movies_df['transcription'] = movies_df['transcription'].apply(normalize_text)

output_csv_file = 'normalized_movies.csv'
movies_df.to_csv(output_csv_file, index=False)
```

## 4. Extracting key phrases and topics
([Go to top](#Capstone-8:-Bringing-It-All-Together))

Use this section to extract the key phrases and topics from the videos.

```python
# Key phrases extraction

from sklearn.feature_extraction.text import TfidfVectorizer

csv_file = 'normalized_movies.csv'
normalized_movies_df = pd.read_csv(csv_file)

# defining the corpus (list of transcriptions)
transcriptions_list = normalized_movies_df['transcription'].tolist()

vectorizer = TfidfVectorizer(ngram_range=(1, 4), max_features=10, stop_words='english')

def extract_key_phrases(text):
    X = vectorizer.fit_transform([text])

    # feature names-words/phrases
    feature_names = vectorizer.get_feature_names_out()

    # tF-iDF values-word/phrase
    dense_matrix = X.todense()
    dense_matrix = dense_matrix.tolist()

    # key phrases
    tfidf_scores = [(feature_names[i], dense_matrix[0][i]) for i in range(len(dense_matrix[0])) if dense_matrix[0][i] > 0]
    tfidf_scores = sorted(tfidf_scores, key=lambda x: x[1], reverse=True)  # Sort by score
    
    return [phrase for phrase, score in tfidf_scores]

# applying the function to extract key phrases for each transcription by row by row
normalized_movies_df['key_phrases'] = normalized_movies_df['transcription'].apply(lambda x: extract_key_phrases(str(x)))

for index, row in normalized_movies_df.iterrows():
    print(f"Document {index+1}: {row['key_phrases']}")

output_csv_file = 'keyphrased_tf.csv'
normalized_movies_df.to_csv(output_csv_file, index=False)

# normalized_movies_df.key_phrases[3]
```

```python
# Topic Extraction

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

def lda():
    corpus = normalized_movies_df['transcription'].tolist()
    
    # TfidfVectorizer to convert each document into a vector.
    vectorizer = CountVectorizer(ngram_range=(1, 4), max_features=5000, stop_words='english')
    X = vectorizer.fit_transform(corpus)
    
    if X.shape[1] == 0:
        raise ValueError("The vocabulary is empty. Please check the corpus and preprocessing steps.")
    
    # LDA model
    lda_model = LatentDirichletAllocation(n_components=20)
    lda_model.fit(X)
    
    # top words
    n_top_words = 10
    feature_names = vectorizer.get_feature_names_out()
    topics = {}
    for topic_idx, topic in enumerate(lda_model.components_):
        top_words_idx = topic.argsort()[:-n_top_words - 1:-1]
        top_words = [feature_names[i] for i in top_words_idx]
        topics[topic_idx] = top_words
    
    # topic distribution
    doc_topic_dist = lda_model.transform(X)
    normalized_movies_df['assigned_topic'] = doc_topic_dist.argmax(axis=1)
    
    print("Global Topics:")
    for idx, words in topics.items():
        print(f"Topic {idx}: {words}")
        
    return lda_model, vectorizer, X, topics

lda_model, vectorizer, X, topics = lda()
```

```python
# Topic assignment

def assign_topic_from_transcription_tfidf(transcription, topics, vectorizer):
    tfidf_vector = vectorizer.transform([transcription])
    # dense array for easier access
    tfidf_dense = tfidf_vector.toarray()[0]
    
    word2idx = vectorizer.vocabulary_
    
    topic_scores = {}
    for topic, keywords in topics.items():
        score = 0.0
        for keyword in keywords:
            if keyword in word2idx:
                idx = word2idx[keyword]
                score += tfidf_dense[idx]
        topic_scores[topic] = score
    # choose highest score
    assigned_topic = max(topic_scores, key=topic_scores.get)
    return assigned_topic

def assign_topics_to_videos(normalized_movies_df, topics, vectorizer):
    normalized_movies_df['assigned_topic'] = normalized_movies_df['transcription'].apply(
        lambda x: assign_topic_from_transcription_tfidf(x, topics, vectorizer)
    )
    
    normalized_movies_df[['video', 'transcription', 'assigned_topic']].to_csv("video_topics.csv", index=False)

    return normalized_movies_df

normalized_movies_df = assign_topics_to_videos(normalized_movies_df, topics, vectorizer)

print(normalized_movies_df[['video', 'transcription', 'assigned_topic']].head(47))
```

## 5. Creating the dashboard
([Go to top](#Capstone-8:-Bringing-It-All-Together))

Use this section to create the dashboard for your solution.

```python
def search_videos(query, topics, vectorizer, normalized_movies_df):
    # topic and keyphrases from query.
    query_topic = assign_topic_from_transcription_tfidf(query, topics, vectorizer)
    query_keyphrases = extract_key_phrases(query)
    
    scores = []
    # separate scores computation for topic match and keyphrase overlap
    for idx, row in normalized_movies_df.iterrows():
        video_topic = row['assigned_topic']
        video_keyphrases = row['key_phrases']
        
        topic_score = 1 if video_topic == query_topic else 0
        
        # fraction of query keyphrases found in the video's keyphrases.
        if query_keyphrases:
            overlap = len(set(query_keyphrases) & set(video_keyphrases))
            keyphrase_score = overlap / len(query_keyphrases)
        else:
            keyphrase_score = 0
        
        composite_score = max(topic_score, keyphrase_score)
        scores.append(composite_score)
    
    normalized_movies_df['composite_score'] = scores
    results = normalized_movies_df.sort_values(by='composite_score', ascending=False)
    
    results = results[results['composite_score'] >= 0.8]
    return results

search_query = "computer vision" # manual search
filtered_videos = search_videos(search_query, topics, vectorizer, normalized_movies_df)
print(filtered_videos[['video', 'assigned_topic', 'key_phrases', 'composite_score']].head(10))
```

```python
def gradio_search(query):
    try:
        # Use the composite search function
        filtered_videos = search_videos(query, topics, vectorizer, normalized_movies_df)
        
        if filtered_videos.empty:
            return f"<h3>No videos found for query: {query}</h3>"
        
        # Build an HTML snippet to display the videos with details, composite score, and video link.
        html_output = "<h3>Search Results:</h3>"
        for index, row in filtered_videos.iterrows():
            video_name = row["video"]
            assigned_topic = row["assigned_topic"]
            composite_score = row["composite_score"]
            s3_url = row.get("s3_url", "#")
            
            html_output += f"""
            <div style="margin-bottom:20px; border:1px solid #ccc; padding:10px;">
                <p><strong>Video:</strong> {video_name}</p>
                <p><strong>Topic:</strong> {assigned_topic}</p>
                <video width="320" height="240" controls>
                    <source src="{s3_url}" type="video/mp4">
                    Your browser does not support the video tag.
                </video>
                <p><a href="{s3_url}" target="_blank">Watch Full Screen</a></p>
            </div>
            """
        return html_output

    except Exception as e:
        return f"<h3>Error: {e}</h3>"
```

```python
import gradio as gr

interface = gr.Interface(
    fn=gradio_search,
    inputs=gr.Textbox(label="Enter Topic or Key Phrases"),
    outputs=gr.HTML(label="Filtered Videos (with Video Links)"),
    title="Searchable Dashboard",
    description="Find course videos by entering key phrases or topics. The results include an embedded video player and a clickable video link."
)

interface.launch(share=True)
```
